## Cleaning and Standardising Column Names

The first step in cleaning after inspecting the raw data sets is to standardise the column names. Having consistent column names makes it easier to write code, merge datasets, run queries, and avoid errors due to spaces, capital letters, or special characters. In this step, all column names are converted to lower case, extra spaces are removed and spaces are replaced by underscores


```python
# function to clean column names

def clean_column_names(df):
    df = df.copy()
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(" ", "_", regex=False)
        .str.replace("-", "_", regex=False)
        .str.replace("/", "_", regex=False)
    )
    return df

# apply column name cleaning to all datasets

app_events_clean = clean_column_names(app_events)
complaints_clean = clean_column_names(complaints)
customers_clean = clean_column_names(customers)
deliveries_clean = clean_column_names(deliveries)
drivers_clean = clean_column_names(drivers)
hubs_clean = clean_column_names(hubs)
incidents_clean = clean_column_names(incidents)
orders_clean = clean_column_names(orders)
vehicles_clean = clean_column_names(vehicles)

print("Column names cleaned successfully.")
```

```python
# store cleaned datasets in a dictionary

cleaned_datasets = {
    "app_events_clean": app_events_clean,
    "complaints_clean": complaints_clean,
    "customers_clean": customers_clean,
    "deliveries_clean": deliveries_clean,
    "drivers_clean": drivers_clean,
    "hubs_clean": hubs_clean,
    "incidents_clean": incidents_clean,
    "orders_clean": orders_clean,
    "vehicles_clean": vehicles_clean
}

# display cleaned column names

for name, df in cleaned_datasets.items():
    print("=" * 80)
    print(f"Dataset: {name}")
    print("=" * 80)
    print(df.columns.tolist())
    print()
```

## Cleaning Date and Time Columns

Several of the NorthStar datasets have date/time columns. These columns need to be converted from text format into proper datetime format for time-based analysis. This step is important for later analysis where we need to do calculations such as delivery duration, complaint trends over time, incident resolution time, customer signup patterns, and app event activity. Using the right date and time format for your columns will help you to get better accuracy in your analysis


```python
# convert date and time columns into datetime format

app_events_clean["event_timestamp"] = pd.to_datetime(app_events_clean["event_timestamp"], errors="coerce")

complaints_clean["created_at"] = pd.to_datetime(complaints_clean["created_at"], errors="coerce")

customers_clean["signup_date"] = pd.to_datetime(customers_clean["signup_date"], errors="coerce")

deliveries_clean["dispatch_time"] = pd.to_datetime(deliveries_clean["dispatch_time"], errors="coerce")
deliveries_clean["delivery_completed_at"] = pd.to_datetime(deliveries_clean["delivery_completed_at"], errors="coerce")

incidents_clean["reported_at"] = pd.to_datetime(incidents_clean["reported_at"], errors="coerce")

orders_clean["order_created_at"] = pd.to_datetime(orders_clean["order_created_at"], errors="coerce")

vehicles_clean["commission_date"] = pd.to_datetime(vehicles_clean["commission_date"], errors="coerce")

print("Date and time columns converted successfully.")
```

## Standardising Text and Categorical Values

The next step after transforming date and time columns is to normalise text-based categorical values. Some columns have values like zones, service types, statuses, channels, incident types, vehicle categories etc. If these values are entered with inconsistent capitalisation or spacing, they can have an impact on grouping, filtering, visualisation and subsequent database queries. This step cleans text values by removing extra spaces and making categorical values consistently Title Case. This improves the reliability of subsequent analysis including complaints by zone, delivery status by hub, incidents by severity and orders by service type


```python
# function to standardise text values in object/string columns

def standardise_text_columns(df):
    df = df.copy()

    text_columns = df.select_dtypes(include=["object"]).columns

    for col in text_columns:
        df[col] = df[col].astype(str).str.strip()
        df[col] = df[col].replace("nan", np.nan)

    return df

# apply text standardisation to all cleaned datasets

app_events_clean = standardise_text_columns(app_events_clean)
complaints_clean = standardise_text_columns(complaints_clean)
customers_clean = standardise_text_columns(customers_clean)
deliveries_clean = standardise_text_columns(deliveries_clean)
drivers_clean = standardise_text_columns(drivers_clean)
hubs_clean = standardise_text_columns(hubs_clean)
incidents_clean = standardise_text_columns(incidents_clean)
orders_clean = standardise_text_columns(orders_clean)
vehicles_clean = standardise_text_columns(vehicles_clean)

print("Text columns standardised successfully.")
```

```python
# standardise zone-related columns

zone_columns = [
    (app_events_clean, "zone_context"),
    (customers_clean, "home_zone"),
    (drivers_clean, "base_zone"),
    (hubs_clean, "zone"),
    (orders_clean, "pickup_zone"),
    (orders_clean, "dropoff_zone"),
    (vehicles_clean, "assigned_zone")
]

for df, col in zone_columns:
    if col in df.columns:
        df[col] = df[col].str.title()

print("Zone-related columns standardised successfully.")
```

## Checking and Handling Missing Values

Analysis reliability may be impacted by missing values, particularly when computing averages, totals, rates, delivery times, complaint trends, and operational performance metrics. All NorthStar datasets are examined for missing values prior to cleaning. This step's goal is to locate missing values and handle them properly based on each column's meaning rather than removing data carelessly. While some missing data might be fine, others might need to be filled in, marked, or eliminated if they have an impact on important analysis


```python
# update cleaned datasets dictionary after previous cleaning steps

cleaned_datasets = {
    "app_events_clean": app_events_clean,
    "complaints_clean": complaints_clean,
    "customers_clean": customers_clean,
    "deliveries_clean": deliveries_clean,
    "drivers_clean": drivers_clean,
    "hubs_clean": hubs_clean,
    "incidents_clean": incidents_clean,
    "orders_clean": orders_clean,
    "vehicles_clean": vehicles_clean
}

# check missing values in each cleaned dataset

for name, df in cleaned_datasets.items():
    print("=" * 80)
    print(f"Missing values in {name}")
    print("=" * 80)

    missing_values = df.isnull().sum()
    missing_values = missing_values[missing_values > 0]

    if missing_values.empty:
        print("No missing values found.")
    else:
        print(missing_values)

    print()
```

```python
# calculate missing value percentage for each dataset

for name, df in cleaned_datasets.items():
    print("=" * 80)
    print(f"Missing value percentage in {name}")
    print("=" * 80)

    missing_percent = (df.isnull().sum() / len(df)) * 100
    missing_percent = missing_percent[missing_percent > 0].round(2)

    if missing_percent.empty:
        print("No missing values found.")
    else:
        print(missing_percent)

    print()
```

### Missing Value Handling Strategy

The type and significance of each column determined the missing value handling. Because they are necessary for connecting datasets, significant identifier columns were kept. 'Unknown' was used to substitute missing values in categorical columns where the lack of information still had analytical relevance. Since the median is less impacted by extreme numbers than the mean, it was used to fill in missing values for numerical columns when suitable. Since replacing missing timestamps with false dates could result in time-based analysis that is misleading, date and time columns were not force-filled

```python
# function to handle missing values carefully

def handle_missing_values(df):
    df = df.copy()

    for col in df.columns:
        if df[col].dtype == "object":
            df[col] = df[col].fillna("Unknown")
        elif pd.api.types.is_numeric_dtype(df[col]):
            if df[col].isnull().sum() > 0:
                df[col] = df[col].fillna(df[col].median())

    return df

# apply missing value handling to all cleaned datasets

app_events_clean = handle_missing_values(app_events_clean)
complaints_clean = handle_missing_values(complaints_clean)
customers_clean = handle_missing_values(customers_clean)
deliveries_clean = handle_missing_values(deliveries_clean)
drivers_clean = handle_missing_values(drivers_clean)
hubs_clean = handle_missing_values(hubs_clean)
incidents_clean = handle_missing_values(incidents_clean)
orders_clean = handle_missing_values(orders_clean)
vehicles_clean = handle_missing_values(vehicles_clean)

print("Missing values handled successfully.")
```

```python
# verify remaining missing values after handling

cleaned_datasets = {
    "app_events_clean": app_events_clean,
    "complaints_clean": complaints_clean,
    "customers_clean": customers_clean,
    "deliveries_clean": deliveries_clean,
    "drivers_clean": drivers_clean,
    "hubs_clean": hubs_clean,
    "incidents_clean": incidents_clean,
    "orders_clean": orders_clean,
    "vehicles_clean": vehicles_clean
}

for name, df in cleaned_datasets.items():
    print("=" * 80)
    print(f"Remaining missing values in {name}")
    print("=" * 80)

    remaining_missing = df.isnull().sum()
    remaining_missing = remaining_missing[remaining_missing > 0]

    if remaining_missing.empty:
        print("No remaining missing values.")
    else:
        print(remaining_missing)

    print()
```

## Checking Key ID Columns and Dataset Relationships

The NorthStar dataset consists of multiple connected files. It's crucial to confirm that key identifier columns can be used to properly connect the data before building integrated datasets. Customers related to orders, orders linked to deliveries and complaints, deliveries linked to drivers, vehicles, hubs, and incidents are all examples of common linkages that are checked in this stage. These verifications help in ensuring that the datasets can be safely combined for additional research


```python
# function to check whether foreign key values exist in the related parent table

def check_relationship(child_df, child_key, parent_df, parent_key, relationship_name):
    total_records = len(child_df)
    matched_records = child_df[child_key].isin(parent_df[parent_key]).sum()
    unmatched_records = total_records - matched_records

    print("=" * 80)
    print(relationship_name)
    print("=" * 80)
    print(f"Total records checked: {total_records}")
    print(f"Matched records: {matched_records}")
    print(f"Unmatched records: {unmatched_records}")
    print(f"Match percentage: {(matched_records / total_records) * 100:.2f}%")
    print()

# relationship checks

check_relationship(orders_clean, "customer_id", customers_clean, "customer_id", "Orders linked to Customers")
check_relationship(complaints_clean, "customer_id", customers_clean, "customer_id", "Complaints linked to Customers")
check_relationship(complaints_clean, "order_id", orders_clean, "order_id", "Complaints linked to Orders")
check_relationship(deliveries_clean, "order_id", orders_clean, "order_id", "Deliveries linked to Orders")
check_relationship(deliveries_clean, "driver_id", drivers_clean, "driver_id", "Deliveries linked to Drivers")
check_relationship(deliveries_clean, "vehicle_id", vehicles_clean, "vehicle_id", "Deliveries linked to Vehicles")
check_relationship(deliveries_clean, "hub_id", hubs_clean, "hub_id", "Deliveries linked to Hubs")
check_relationship(incidents_clean, "delivery_id", deliveries_clean, "delivery_id", "Incidents linked to Deliveries")
check_relationship(app_events_clean, "customer_id", customers_clean, "customer_id", "App Events linked to Customers")
check_relationship(app_events_clean, "order_id", orders_clean, "order_id", "App Events linked to Orders")
```

## Creating Derived and Integrated Datasets

Following the cleaning of the individual files, extra processed datasets are generated to help in further analysis. These derived datasets integrate relevant NorthStar records to provide more helpful analytical views. The linked datasets are used to assess customer complaints, delivery performance, hub efficiency, driver problems, and overall operational performance across the organization

```python
# create an integrated orders and deliveries dataset

integrated_orders_deliveries = orders_clean.merge(
    deliveries_clean,
    on="order_id",
    how="left"
)

print("integrated_orders_deliveries created successfully.")
print(integrated_orders_deliveries.shape)
display(integrated_orders_deliveries.head())
```

```python
# create customer complaint summary dataset

customer_complaint_summary = complaints_clean.groupby("customer_id").agg(
    total_complaints=("complaint_id", "count"),
    average_resolution_days=("resolution_days", "mean"),
    total_compensation=("compensation_amount", "sum")
).reset_index()

print("customer_complaint_summary created successfully.")
print(customer_complaint_summary.shape)
display(customer_complaint_summary.head())
```

```python
# create driver incident summary dataset

driver_incident_data = deliveries_clean.merge(
    incidents_clean,
    on="delivery_id",
    how="left"
)

driver_incident_summary = driver_incident_data.groupby("driver_id").agg(
    total_deliveries=("delivery_id", "count"),
    total_incidents=("incident_id", "count"),
    average_resolved_hours=("resolved_hours", "mean")
).reset_index()

driver_incident_summary["incident_rate"] = (
    driver_incident_summary["total_incidents"] / driver_incident_summary["total_deliveries"]
).round(3)

print("driver_incident_summary created successfully.")
print(driver_incident_summary.shape)
display(driver_incident_summary.head())
```

```python
# create hub performance summary dataset

hub_delivery_data = deliveries_clean.merge(
    hubs_clean,
    on="hub_id",
    how="left"
)

hub_performance_summary = hub_delivery_data.groupby(["hub_id", "hub_name", "zone", "hub_type"]).agg(
    total_deliveries=("delivery_id", "count"),
    average_route_distance_km=("route_distance_km", "mean"),
    average_route_overrides=("manual_route_override_count", "mean"),
    missing_proof_count=("proof_of_completion_missing", "sum"),
    average_customer_rating=("customer_rating_post_delivery", "mean"),
    total_fuel_or_charge_cost=("fuel_or_charge_cost", "sum")
).reset_index()

print("hub_performance_summary created successfully.")
print(hub_performance_summary.shape)
display(hub_performance_summary.head())
```

```python
# create a master analysis dataset by combining key operational data

complaint_order_summary = complaints_clean.groupby("order_id").agg(
    complaint_count=("complaint_id", "count"),
    average_resolution_days=("resolution_days", "mean"),
    total_compensation=("compensation_amount", "sum")
).reset_index()

incident_delivery_summary = incidents_clean.groupby("delivery_id").agg(
    incident_count=("incident_id", "count"),
    average_resolved_hours=("resolved_hours", "mean")
).reset_index()

northstar_master_analysis = orders_clean.merge(customers_clean, on="customer_id", how="left")
northstar_master_analysis = northstar_master_analysis.merge(deliveries_clean, on="order_id", how="left")
northstar_master_analysis = northstar_master_analysis.merge(hubs_clean, on="hub_id", how="left")
northstar_master_analysis = northstar_master_analysis.merge(vehicles_clean, on="vehicle_id", how="left")
northstar_master_analysis = northstar_master_analysis.merge(complaint_order_summary, on="order_id", how="left")
northstar_master_analysis = northstar_master_analysis.merge(incident_delivery_summary, on="delivery_id", how="left")

northstar_master_analysis["complaint_count"] = northstar_master_analysis["complaint_count"].fillna(0)
northstar_master_analysis["incident_count"] = northstar_master_analysis["incident_count"].fillna(0)
northstar_master_analysis["total_compensation"] = northstar_master_analysis["total_compensation"].fillna(0)

print("northstar_master_analysis created successfully.")
print(northstar_master_analysis.shape)
display(northstar_master_analysis.head())
```
